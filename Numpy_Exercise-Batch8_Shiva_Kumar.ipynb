#### 21. Compute the average and strike rate for each player and sort it decending way (seperately for avg and strike rate)

```python
# Solution 
import numpy as np
runs=np.array([[54,60,10,65],[102,2,78,32],[3,110,1,92],[56,6,12,42],[12,8,22,11]])
print(np.sort(runs.mean(1))[::-1])#Avg

balls=np.array([[50,42,22,65],[95,8,65,43],[9,100,7,101],[27,3,11,31],[24,20,36,27]])
totalruns=runs.sum(1)
totalballs=balls.sum(1)
print(np.sort((totalruns/totalballs)*100)[::-1])#strike rate
```

#### 23. The class incharge decided to give a mercy mark of 5 for those who scored between 45 and 49. Also the bonus of 5 marks awarded to the students who scored more than 80. Print all the student marks and compute the increase in the pass percentage of class. Any score more than 100 must be recorded as only 100.

```python
# Solution
import numpy as np
marks = np.array([65,84,47,56,89,98,48,77,80,39])

inipass=len(marks[marks>=50])/len(marks)*100
print(inipass)             

condlist = [(marks>=45)&(marks<=49), (marks>80)&(marks<=95), (marks>96)&(marks<=100)]
choicelist = [marks+5, marks+5,100]

marks = np.select(condlist, choicelist, default=marks)
marks

finpass=len(marks[marks>=50])/len(marks)*100
print(finpass)

#20% increase in pass percentage
```
